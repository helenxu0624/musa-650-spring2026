```python
from pystac_client import Client
# use pystac-client to open up our STAC API
catalog = Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1",
    modifier=planetary_computer.sign_inplace,
)
catalog.title
# use the STAC API to search for assets meeting some criteria
# all_collections = [i.id for i in catalog.get_collections()]
# sentinel_collections = [
#     collection for collection in all_collections if "sentinel" in collection
# ]
# sentinel_collections
# define search parameters
bbox = [8.847198, 40.193395, 8.938865, 40.241895]  # Region over a lake in Nevada, USA
# Before the fire:
pre_f = "2025-06-03"
# After the fire:
post_f = "2025-06-21"
collection = "sentinel-2-l2a"
```

```python
items = list(
    catalog.search(
        bbox=bbox,
        #  for June 3 only
        datetime=f"{pre_f}T00:00:00Z/{pre_f}T23:59:59Z",
        collections="sentinel-2-l2a",
    ).item_collection()
)

print(f"Total Items Found for Sentinel-2 L2A over Sardinia: {len(items)}")
```
